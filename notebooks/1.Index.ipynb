# Files

## Data preprocessing

* [Trial data preprocessing](Preprocessing-TrialData.ipynb)

Processes the files containing trial by trial data from all subjects during an experiment.

* [Self reported answers and metrics preprocessing](Preprocessing-SelfReported.ipynb)

Processes self reported answers dividing it by task and subject. The data is then stacked with selection and accuracy metrics obtained from the trial data.

* [Stack self reported answers and metrics](Preprocessing-StackData.ipynb)

Stacks self reported and metrics arrays split in phases to turn them into a single unified array.


## Pilot examples

* [Task selection, error and LP](Selection-Error-LP.ipynb)

Plots of cumulative task selection, error rate during the experiment and learning progress based on the error rate (interactive version).

* [Task selection, error and LP (Static)](Selection-Error-LP-Static.ipynb)

Plots of cumulative task selection, error rate during the experiment and learning progress based on the error rate (static version).

* [Probability of switching to another task given LP](TaskSwitching-givenLP.ipynb)

Shows a range of LP values and the probability subjects will switch to a different task if they have certain LP (interactive version).

* [Probability of switching to another task given LP (Static)](TaskSwitching-givenLP-Static.ipynb)

Shows a range of LP values and the probability subjects will switch to a different task if they have certain LP (static version).


## Selection comparison among experimental groups


* [Selection comparison](Selection-Comparison-FreeExplorationGroups.ipynb)

Shows various comparisons among the free exploration and the free exploration with training groups. Cumulative action selection during the exploration phase, histograms showing number of times people are switching, how many times people return to a task after they have abandoned it and selection dynamics on 1D and R tasks.

* [Selection comparison](Selection-Comparison.ipynb)

Extends the comparison to all groups.

* [Final performance and selection comparison](Performance-Selection-Comparison.ipynb)

Comparison among all experimental groups (free exploration, informed free exploration with training, uninformed free exploration with training, informed strategic and uninformed strategic) on task final performance and task selection.


## Analysis of self reported factors

* [Self reported scores in tasks in Strategic Learning groups](Analysis-SelfReported-Strategic.ipynb)

Relationships between tasks and self reported scores. It also inclues correlations between various of these factors. This file includes data from the Strategic Learning groups.

* [Self reported scores in tasks in Free Exploration groups](Analysis-SelfReported-FreeExploration.ipynb)

Relationships between tasks and self reported scores. It also inclues correlations between various of these factors. This file includes data from the Free Exploration and the Free Exploration with Training groups.


## Other files

* [Task switching according to LP](Strategic-TaskSwitchLP.ipynb)

Similarly to an earlier file, this one takes a look at the probability of switching to another task given certain value of LP. It also takes a closer look at what the probabilities are when LP=0 on each task and what LP values are most likely to lead to switch task.


* [Open Loop](OpenLoop.ipynb)

This analysis considers the relation between the training stage and the first trial selection. For example, accuracy during training, relative accuracy, LP, relative LP and order of presentation.

* [Final performance and selection comparison](Feature-relative-importance.ipynb)

Initial attempt at establishing the influence of the factors in task selection. It includes an example of using decision trees to choose features for partitions (entropy) and another example using PCA and a regressor.
